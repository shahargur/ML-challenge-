```python
#convert all valuse from int to float

url= data['url'] 
url= pd.DataFrame(url)

data = data.drop(columns= 'url')

features= data.columns

for f in features:
    data[f] = data[f].astype(float)
```

```python
#data preprocessing for the feature importance analysis
shares= data[' shares'] 
shares= pd.DataFrame(shares)

X = data.drop(columns= ' shares')
y = shares
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
#normalize for PCA
ss = StandardScaler()
X_train_scaled = ss.fit_transform(X)

pca = PCA().fit(X_train_scaled)

pp.plot(pca.explained_variance_ratio_.cumsum(), lw=3, color='#087E8B')
pp.title('Cumulative explained variance by number of principal components', size=15)
pp.show()
```

```python
# decision tree for feature importance on a regression problem
from sklearn.datasets import make_regression
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(X, y)
importance = model.feature_importances_
low_importance_features=[]

for i,v in enumerate(importance):
    print('Feature: %0d: %s, Score: %.5f' % (i,features[i],v))
    if v<0.005:
        low_importance_features= low_importance_features+[features[i]]
pp.bar([x for x in range(len(importance))], importance)
pp.show()
print('the low importance features are:\n')
low_importance_features
```

```python
#remove the low importance features
X = X.drop(columns= low_importance_features)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn import tree
from sklearn.svm import SVC
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score


interval_transformer = preprocessing.KBinsDiscretizer(n_bins= 5, encode='ordinal')
interval= interval_transformer.fit_transform(y)
classified_shares = pd.DataFrame(interval, columns=[' shares'])

X_train, X_test, y_train, y_test = train_test_split(X, classified_shares.values.ravel(),
                                                    test_size=0.3,
                                                    random_state=5)
```

```python
#normlize the data
sc= StandardScaler()
X_train= sc.fit_transform(X_train)
X_test= sc.transform(X_test)
```

```python
#Decision Trees
#choose optimal hyper parameters
model_DT= tree.DecisionTreeClassifier()
hyper_parameter_DT = {'criterion':['gini','entropy'],'max_depth': np.arange(5,30)}

grid_search = GridSearchCV(estimator= model_DT,
                           param_grid= hyper_parameter_DT,
                          scoring= 'accuracy',
                          cv=10,
                          n_jobs= -1)
grid_search= grid_search.fit(X_train, y_train)
    
print("The best hyper parameters found are:")
print(grid_search.best_params_)
print("with accuracy:")
print(grid_search.best_score_)
```

```python
#set the model with the best parameters and evaluate the model
model_DT= tree.DecisionTreeClassifier(criterion= 'gini', max_depth= 7)
model_DT.fit(X_train, y_train)
y_pred= model_DT.predict(X_test)
cm= confusion_matrix(y_test, y_pred)
accuracy= accuracy_score(y_test, y_pred)
print("accuracy of the DT is:")
print(accuracy)
```

```python
#Support Vector Machines
#choose optimal hyper parameters
if __name__ == "__main__":
    model_svm= SVC()
    hyper_parameter_svm= [{'C': [1, 10, 100, 1000], 'kernel': ['linear']},
                          {'C': [1, 10, 100, 1000], 'kernel': ['rbf'],'gamma': [1, 0.1, 0.01, 0.001, 0.0001]}]
    grid_search= GridSearchCV(estimator= model_svm,
                          param_grid= hyper_parameter_svm,
                          scoring= 'accuracy',
                          cv=5,
                          n_jobs= -1,
                          verbose = 10)
    grid_search= grid_search.fit(X_train, y_train)
    print("The best hyper parameters found are:")
    print(grid_search.best_params_)
    print("with accuracy:")
```

```python
#set the model with the best parameters and evaluate the model
model_svm= SVC(C=100, gamma=0.001, kernel='rbf')
model_svm.fit(X_train, y_train)
y_pred= model_svm.predict(X_test)
cm= confusion_matrix(y_test, y_pred)
accuracy= accuracy_score(y_test, y_pred)
print("accuracy of the SVM is:")
print(accuracy)
```

```python
#An ensamble methodology - Adaboost
#choose optimal hyper parameters
from sklearn.ensemble import AdaBoostClassifier

if __name__ == "__main__":
    hyper_parameter_Adaboost= {'n_estimators':[500,1000,2000],'learning_rate':[.001,0.01,.1]}

    grid_search= GridSearchCV(estimator= AdaBoostClassifier(),
                              param_grid= hyper_parameter_Adaboost,
                              scoring= 'accuracy',
                              cv=5,
                              n_jobs= -1)
    grid_search= grid_search.fit(X_train, y_train)

    print("The best hyper parameters found are:")
    print(grid_search.best_params_)
    print("with accuracy:")
    print(grid_search.best_score_)
```

```python
#set the model with the best parameters and evaluate the model
from sklearn.ensemble import AdaBoostClassifier
model_ABC= AdaBoostClassifier(learning_rate= 0.1, n_estimators= 1000)
model_ABC.fit(X_train, y_train)
y_pred= model_ABC.predict(X_test)
cm= confusion_matrix(y_test, y_pred)
accuracy= accuracy_score(y_test, y_pred)
print("accuracy of the Adaboost classifier is:")
print(accuracy)
```

```python
#MLPNs
#choose optimal hyper parameters
from sklearn.neural_network import MLPClassifier
if __name__ == "__main__":
    hyper_parameter_MLP= {
        'hidden_layer_sizes': [(10,30,10),(20,)],
        'activation': ['tanh', 'relu'],
        'solver': ['sgd', 'adam'],
        'alpha': [0.0001, 0.05],
        'learning_rate': ['constant','adaptive'],
    }

    grid_search= GridSearchCV(estimator= MLPClassifier(max_iter=100),
                               param_grid= hyper_parameter_MLP,
                              scoring= 'accuracy',
                              cv=5,
                              n_jobs= -1)
    grid_search= grid_search.fit(X_train, y_train)

    print("The best hyper parameters found are:")
    print(grid_search.best_params_)
    print("with accuracy:")
    print(grid_search.best_score_)
```

Forth phrase – Model Selection
First of all, I needed to prepper the data for a classification task. I did it by spliting  the 'shares' column into 5 classes according to ranges. 
After that I could split the data into training set and testing set. I did it using train_test_split which perform random split. I chose to split the data in an acceptable division of 70% training and 30% testing.

I normalize the data using StandardScaler. The standard score of a sample x is calculated as: 
z = (x - u) / s
Where u is the mean of the training samples.
I could use this function on all the features because I saw that in the feature selection phase all the Boolean features were dropped. We obtain with numerical feature only.
This part is extremely important. When we look at the original data set we can see that the ranges of values of the features are very different from each other, which can lead to the construction of a poor model.
Now when the data is ready I could build the models. 
For each model I used GridSearchCV function in order to find the optimal parameters to build my model with. The parameters of the model used to apply these methods are optimized by cross-validated grid-search over a parameter grid. 
I set the n_jobs to -1, which means the function will using all processors. multi-processing requires grid search to run in a "__main__" protected block.
